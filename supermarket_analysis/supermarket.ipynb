<!-- Feature Engineering Time-Based Section --> 
<div style="background-color:#FAFAF5; padding: 20px; border-left: 6px solid #D97706; border-radius: 10px; margin-bottom: 20px;"> <h2 style="color:#92400E;"><strong>Let's Do Some Feature Engineering</strong></h2> <h3 style="color:#B45309;"><strong>Uncover Time-Based Patterns in Customer Behavior</strong></h3> <p style="color: #78350F; font-size: 15px; line-height: 1.6;"> The <code>Date</code> and <code>Time</code> columns hold valuable information about customer behavior over time. We can extract new features to enrich our analysis: </p> <ul style="color: #78350F; font-size: 15px;"> <li><code>Day</code>, <code>Month</code>, <code>Hour</code> — provide temporal context</li> <li><code>Weekday</code> — helps capture weekly purchasing trends</li> <li><code>Time_slot</code> — maps hour to general periods (e.g., Morning, Afternoon)</li> </ul> <br> <table style="width:100%; border-collapse: collapse; font-size: 15px;"> <thead style="background-color:#92400E;"> <tr> <th style="padding: 10px; text-align: left; border-bottom: 1px solid #FCD34D;">Goal</th> <th style="padding: 10px; text-align: left; border-bottom: 1px solid #FCD34D;">Feature</th> </tr> </thead> <tbody style="color: #92400E;"> <tr> <td>Extract time context</td> <td><code>Day</code>, <code>Month</code>, <code>Hour</code></td> </tr> <tr> <td>Capture weekly patterns</td> <td><code>Weekday</code></td> </tr> <tr> <td>Create interpretable time slots</td> <td><code>Time_slot</code></td> </tr> </tbody> </table> </div>

```python
df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Weekday'] = df['Date'].dt.day_name()
```

```python
df['Hour'] = df['Time'].dt.hour

# Define time slots
def time_slot(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

df['Time_slot'] = df['Hour'].apply(time_slot)
```

<!-- Time-Based EDA Section --> 
<div style="background-color:#EFF6FF; padding: 20px; border-left: 6px solid #3B82F6; border-radius: 10px; margin-bottom: 20px;"> <h3 style="color: #1D4ED8;"><strong>Time-Based Order Analysis</strong></h3> <p style="color: #1E3A8A; font-size: 15px;">Let’s explore how orders vary by different time-related features extracted earlier. These questions guide our investigation into customer purchasing behavior over time:</p> <ol style="color: #1E40AF; font-size: 15px; line-height: 1.8;"> <li> <strong>Order Count by Hour</strong> <br> <em style="color: #3B82F6;">At what time of day do most orders occur?</em> </li> <li> <strong>Order Count by Day of the Month</strong> <br> <em style="color: #3B82F6;">Are there specific dates in the month where order frequency peaks?</em> </li> <li> <strong>Order Count by Month</strong> <br> <em style="color: #3B82F6;">Which months have higher or lower order volumes?</em> </li> <li> <strong>Order Count by Weekday</strong> <br> <em style="color: #3B82F6;">Which days of the week are most popular for orders?</em> </li> <li> <strong>Order Count by Time Slot</strong> <br> <em style="color: #3B82F6;">Do customers order more in the morning, afternoon, or evening?</em> </li> </ol> </div>

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the plot style
sns.set(style="whitegrid")
plt.figure(figsize=(16, 20))

# Plot Hour
plt.subplot(3, 2, 1)
sns.barplot(x=df['Hour'].value_counts().index, y=df['Hour'].value_counts().values, palette="viridis")
plt.title('Order Count by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Count')

# Plot Day
plt.subplot(3, 2, 2)
sns.barplot(x=df['Day'].value_counts().index, y=df['Day'].value_counts().values, palette="plasma")
plt.title('Order Count by Day of Month')
plt.xlabel('Day')
plt.ylabel('Count')

# Plot Month
plt.subplot(3, 2, 3)
sns.barplot(x=df['Month'].value_counts().index, y=df['Month'].value_counts().values, palette="coolwarm")
plt.title('Order Count by Month')
plt.xlabel('Month')
plt.ylabel('Count')

# Plot Weekday
plt.subplot(3, 2, 4)
sns.barplot(x=df['Weekday'].value_counts().index, y=df['Weekday'].value_counts().values, palette="Set2")
plt.title('Order Count by Weekday')
plt.xlabel('Day of the Week')
plt.ylabel('Count')

# Plot Time Slot
plt.subplot(3, 2, 5)
sns.barplot(x=df['Time_slot'].value_counts().index, y=df['Time_slot'].value_counts().values, palette="Set3")
plt.title('Order Count by Time Slot')
plt.xlabel('Time Slot')
plt.ylabel('Count')

# Adjust layout
plt.tight_layout()
plt.show()
```
